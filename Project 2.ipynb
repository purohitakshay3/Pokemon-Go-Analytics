```python
from bs4 import BeautifulSoup
import urllib


in_dict = {}

def get_data(filename):
    #fname = r"C:\Users\Akshay PC\Desktop\DS PROJECT 2\pokemon_5378\data\2016-07-21\00_40_pokemon_android.html"
    fname = filename
#http://stackoverflow.com/questions/32547543/reading-an-html-file-from-folder-in-python
    HtmlFile = open(fname,'r')
    r = HtmlFile.read()
#print r
    soup = BeautifulSoup(r, "lxml")
    soup.prettify()
#one_prof = soup.find('div', { "class" : "tiny-star star-rating-non-editable-container" })

#For Average Rating
    avg_rat = soup.find('div', { "class" : "score" }).get_text()
#https://www.crummy.com/software/BeautifulSoup/bs3/documentation.html
#print("Android Average Rating = {}".format((avg_rat)))

#For Total No. of Ratings
    tot_rat = soup.find('span', { "class" : "reviews-num" }).get_text().replace(",", "")
#http://stackoverflow.com/questions/3559559/how-to-delete-a-character-from-a-string-using-python
#print("Total Rating = {}".format((tot_rat)))

#For Five-Star Rating
    #print("\nNumber of Ratings from 1 - 5")
    five_rat = soup.find('div', { "class" : "rating-bar-container five" })
    five_rating = five_rat.find('span', { "class" : "bar-number" }).get_text().replace(",", "")
#print("Five-star Rating = {}".format((five_rating)))

#For Four-Star Rating
    four_rat = soup.find('div', { "class" : "rating-bar-container four" })
    four_rating = four_rat.find('span', { "class" : "bar-number" }).get_text().replace(",", "")
#print("Four-star Rating = {}".format((four_rating)))

#For Three-Star Rating
    three_rat = soup.find('div', { "class" : "rating-bar-container three" })
    three_rating = three_rat.find('span', { "class" : "bar-number" }).get_text().replace(",", "")
#print("Three-star Rating = {}".format((three_rating)))

#For Two-Star Rating
    two_rat = soup.find('div', { "class" : "rating-bar-container two" })
    two_rating = two_rat.find('span', { "class" : "bar-number" }).get_text().replace(",", "")
#print("Two-star Rating = {}".format((two_rating)))

#For One-Star Rating
    one_rat = soup.find('div', { "class" : "rating-bar-container one" })
    one_rating = one_rat.find('span', { "class" : "bar-number" }).get_text().replace(",", "")
#print("One-star Rating = {}".format((one_rating)))

#For File Size
    file_size = soup.find('div', { "itemprop" : "fileSize" }).get_text()
#print("\nFile size = {}".format(file_size))

#Internal Dictionary
    in_dict['android_current_ratings'] = avg_rat
    in_dict['android_total_ratings'] = tot_rat
    in_dict['android_rating_1'] = one_rating
    in_dict['android_rating_2'] = two_rating
    in_dict['android_rating_3'] = three_rating
    in_dict['android_rating_4'] = four_rating
    in_dict['android_rating_5'] = five_rating
    in_dict['android_file_size'] = file_size[:3]
    
#Desired Output
    #print("\n{}, {}, {}, {}, {}, {}, {},{}").format(avg_rat, tot_rat, five_rating, four_rating, three_rating, two_rating, one_rating, file_size[:3])
```

```python
import os
import glob

out_dict = {}

path = r"C:\Users\Akshay PC\Desktop\DS PROJECT 2\pokemon_5378\data\2016-07-21"
    
#http://stackoverflow.com/questions/18262293/python-open-every-file-in-a-folder

for filename in glob.glob(os.path.join(path, '*android.html')):
    try:
        get_data(filename)
        #print filename
        #print in_dict
        key = "datetime({}, {}, {}, {}, {})".format(filename[-37:-33], filename[-32:-30], filename[-29:-27], filename[-26:-24], filename[-23:-21])
        out_dict[key] = in_dict
        #print key
        #2016-07-28\00_00_pokemon_android.html
    except:
        continue

print out_dict 
```

```python
import pandas as pd

df = pd.DataFrame()
df['Date'] = out_dict.keys()
df['DateValue'] = out_dict.values()
#http://stackoverflow.com/questions/18837262/convert-python-dict-into-a-dataframe
print df
```

```python
import json
#http://stackoverflow.com/questions/16923281/pandas-writing-dataframe-to-json-file
data = df.to_json('data.json')
#with open('data.json', 'w') as fp:
#    json.dump(data, fp)
```

```python
#http://stackoverflow.com/questions/16923281/pandas-writing-dataframe-to-csv-file
for column in df.columns:
    for idx in df[column].index:
        x = df.get_value(idx,column)
        try:
            x = unicode(x.encode('utf-8','ignore'),errors ='ignore') if type(x) == unicode else unicode(str(x),errors='ignore')
            df.set_value(idx,column,x)
        except Exception:
            print 'encoding error: {0} {1}'.format(idx,column)
            df.set_value(idx,column,'')
            continue
            
df.to_csv('data.csv')
df.to_excel('data.xlsx')
```
